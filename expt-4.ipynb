```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
```

```python
data = load_iris()
X = data.data
y = data.target

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
k_values = [3, 5]

results = []
weights_type = ['uniform','distance',lambda d:1/(d + 1e-5)]

for w in weights_type:
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k, weights=w)
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        report_dict = classification_report(
            y_test,
            y_pred,
            target_names=data.target_names,
            output_dict=True
        )
        
        # Convert to DataFrame
        report_df = pd.DataFrame(report_dict).transpose()
        
        results.append({
            "k": k,
            "Report": report_df
        })
```

```python
print("Uniform:-")
for i in range(0,len(k_values)):
    print("K:-",results[i]["k"],"\n",results[i]["Report"],end="\n\n")
print("\nDistance-Weighted:-")
for i in range(2,2+len(k_values)):
    print("K:-",results[i]["k"],"\n",results[i]["Report"],end="\n\n")
print("\nLocally-Weighted:-")
for i in range(4,4+len(k_values)):
    print("K:-",results[i]["k"],"\n",results[i]["Report"],end="\n\n")
```
